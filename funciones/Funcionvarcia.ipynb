```python
import matplotlib.pyplot as plt
```

```python
def FunciónBlanco():
    plt.figure(figsize=(20,4),dpi=600)
    plt.plot()
    plt.legend(fontsize=12)
    plt.title ('Señal', fontsize=20)
    plt.grid (True)
    plt.savefig('imagenes/blanco.jpg', dpi=600, bbox_inches='tight')
    plt.show()
```

```python
FunciónBlanco()
```
