### Función para obtener el espectro de una señal

Esta función permite obtener el espectro de una señal cualquiera. Para ello es necesario ingresar la señal (xinput) cuyo espectro se desea obtener. El vector xinput debe tener dimensión 1xN.

```python
def Espectro(xinput):
    N = len(xinput); p = np.arange(N); espec = np.ndarray(N,dtype=complex)
    for i in np.arange(N):
        espec[i] = np.sum(xinput * np.exp(-2j*np.pi*p*i/N)/N)
    return espec
```

### Función para obtener la señal a partir de su espectro

Esta función devuelve una señal cuyo espectro es el vector xinput. El vector xinput debe tener dimensión 1xN.

```python
def señal(xinput):
    N = len(xinput); n = np.arange(N); señal = np.ndarray(N,dtype=complex)
    for i in np.arange(N):
        señal[i] = np.sum(xinput * np.exp(2j*np.pi*n*i/N))
        A = señal.real
    return A
```
